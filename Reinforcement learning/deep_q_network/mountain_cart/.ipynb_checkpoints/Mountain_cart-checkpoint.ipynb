```python
import gymnasium as gym
import numpy as np
import imageio
from IPython.display import Image
import torch
import torch.nn as nn
import torch.optim as optim
from collections import deque
import matplotlib.pyplot as plt
import random
import copy
from tqdm import tqdm
import pickle
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
class ReplayBuffer:
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = deque(maxlen=capacity)

    def insert(self, state, action, reward, next_state):
        self.memory.append((state, action, reward, next_state))

    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)
```

```python
class QNetwork(nn.Module):
    def __init__(self, input_size, output_size):
        super(QNetwork, self).__init__()
        self.fc1 = nn.Linear(input_size,64)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(64,64)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(64, output_size)

        for param in self.parameters():
            param.requires_grad = True

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        return x
```

```python
def plot_reward(rewards_list):
    episodes = range(0,len(rewards_list))
    plt.plot(episodes, rewards_list)
    plt.title('Rewards over Episodes')
    plt.xlabel('Episodes')
    plt.ylabel('Total Reward')
    plt.show()
```

```python
class DQNAgent:
    def __init__(self, state_size, action_size, gamma=0.99, epsilon_start=1.0):
        self.state_size = state_size
        self.action_size = action_size
        self.gamma = gamma
        self.epsilon = epsilon_start
        self.replay_buffer = ReplayBuffer(capacity=10000)

        self.q_net = QNetwork(state_size, action_size).to(device)
        self.target_net = QNetwork(state_size, action_size).to(device)
        for param in self.target_net.parameters():
              param.requires_grad = False
        self.optimizer = optim.Adam(self.q_net.parameters(), lr=0.001)
        self.criterion = nn.MSELoss()

    def select_action(self, state):
        if np.random.rand() < self.epsilon:
            return np.random.choice(self.action_size)
        with torch.no_grad():
            state = torch.FloatTensor(state).to(device)
            q_values = self.target_net(state)
            return q_values.argmax().item()

    def train_model(self, data_loader):
        states_batch, actions_batch, rewards_batch, next_states_batch = data_loader
        
        states_batch = np.array(states_batch)
        actions_batch = np.array(actions_batch)
        rewards_batch = np.array(rewards_batch)
        next_states_batch = np.array(next_states_batch)
        
        state_tensor = torch.tensor(states_batch, dtype=torch.float32).to(device)
        action_tensor = torch.tensor(actions_batch, dtype=torch.long).to(device)
        reward_tensor = torch.tensor(rewards_batch, dtype=torch.float32).to(device)
        next_state_tensor = torch.tensor(next_states_batch, dtype=torch.float32).to(device)
        
        target = reward_tensor + self.gamma * self.target_net(next_state_tensor).max(1)[0].detach()
        predicted = self.q_net(state_tensor).gather(1, action_tensor.unsqueeze(1)).squeeze(1)
        loss = self.criterion(predicted,target)
        
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
        
        return 
```

```python
env = gym.make('MountainCar-v0',render_mode="rgb_array_list")
state_size = env.observation_space.shape[0]
action_size = env.action_space.n
```

```python
agent = DQNAgent(state_size, action_size)
num_episodes = 400
batch_size = 32
list_rewards = []
for episode in range(num_episodes):
    state , _ = env.reset()
    state = np.float32(state)
    total_reward = 0
    count = 0
    terminated = False
    truncated = False
    while not (terminated or truncated):
        action = agent.select_action(state)
        next_state, reward, terminated, truncated, _ = env.step(action)
        total_reward += reward
        next_state = np.float32(next_state)
        agent.replay_buffer.insert(state, action, reward, next_state)

        if len(agent.replay_buffer.memory) > batch_size:
            transitions = agent.replay_buffer.sample(batch_size)
            data_loader = zip(*transitions)
            agent.train_model(data_loader)
        agent.epsilon = max(0.01,agent.epsilon*0.995)
        if count % 8 == 0:
            agent.target_net = copy.deepcopy(agent.q_net)
        state = next_state
        count += 1
    list_rewards.append(total_reward)
    if episode % 20 == 0:
      print(f"Episode: {episode}, Total Reward: {total_reward}, Epsilon: {agent.epsilon}")
```

<p>observe that agent quickly starts learning once other than -200 reward is detected</p>

```python
plot_reward(list_rewards)
```

## Comparing hyperparameter

```python
with open('b40.pkl', 'rb') as file:
    l40 = pickle.load(file)
with open('b60.pkl', 'rb') as file:
    l60 = pickle.load(file)
with open('b20.pkl', 'rb') as file:
    l20 = pickle.load(file)

plt.plot(l20, label='batch_size = 20')
plt.plot(l40, label='batch_size = 40')
plt.plot(l60, label='batch_size = 60')
plt.legend()
plt.xlabel('Episodes')
plt.ylabel('Total_rewards')
plt.title('Rewards for Various Batch Sizes')
plt.show()

#------------------------------
def smooth(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')

window_size = 25
l40_smooth = smooth(l40, window_size)
l60_smooth = smooth(l60, window_size)
l20_smooth = smooth(l20, window_size)
plt.plot(l20_smooth, label='batch_size = 20')
plt.plot(l40_smooth, label='batch_size = 40')
plt.plot(l60_smooth, label='batch_size = 60')
plt.legend()
plt.xlabel('Episodes')
plt.ylabel('Total_rewards')
plt.title('Smoothed Curves of Rewards for Various Batch Sizes')
plt.show()
```

<p>As  there is a lot of variation in the graph we smoothened over window_size of 25 the graph to interpret the results</p>
<p>we can see from the above graph as the batch size increases the graph will converge more quickly but also note that one other than -200 reward is achieved the agent quickly starts learning but  this can occur at any point of time so the start time when the agent picked up learning is random but one detected we can see batch size of 60 has a sttep jump where as batch size 20 has gradual increase</p>
