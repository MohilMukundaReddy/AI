```python
import gymnasium as gym
import numpy as np
import imageio
import cv2
from IPython.display import Image
import torch
import torch.nn as nn
import torch.optim as optim
from collections import deque
import matplotlib.pyplot as plt
import random
import copy
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
class ReplayBuffer:
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = deque(maxlen=capacity)

    def insert(self, state, action, reward, next_state):
        self.memory.append((state, action, reward, next_state))

    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)
```

```python
def convert(array, new_shape=(84,84)):
    bw_image = np.mean(array, axis=2).astype(np.uint8)
    resized_bw_image = cv2.resize(bw_image, new_shape, interpolation=cv2.INTER_AREA)
    return resized_bw_image/255.0

def update(que,state):
    que.append(convert(state))
    return

def combine(que):
    return np.stack(que)
```

```python
env = gym.make("ALE/Pong-v5")
action_size = env.action_space.n
```

```python
def show_3d_array(array):
    plt.imshow(array)
    plt.axis('off')
    plt.show()

state,_ = env.reset()
show_3d_array(state)
# print(convert(state))
plt.imshow(convert(state), cmap='gray')
```

```python
class QNetwork(nn.Module):
    def __init__(self, output_size):
        super(QNetwork, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=4, out_channels=32, kernel_size=8, stride=4)
        self.relu1 = nn.ReLU()
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=4, stride=2)
        self.relu2 = nn.ReLU()
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=64, kernel_size=5, stride=4)
        self.relu3 = nn.ReLU()
        self.flat = nn.Flatten()
        self.fc1 = nn.Linear(256,128)
        self.fc2 = nn.Linear(128,output_size)

        self.init_weights()
        for param in self.parameters():
            param.requires_grad = True

    def init_weights(self):
        for layer in self.children():
            if isinstance(layer, nn.Conv2d) or isinstance(layer, nn.Linear):
                nn.init.xavier_uniform_(layer.weight)

    def forward(self, x):
        x = self.relu1(self.conv1(x))
        x = self.relu2(self.conv2(x))
        x = self.relu3(self.conv3(x))
        x = self.flat(x)
        x = self.fc1(x)
        x = self.fc2(x)
        return x
```

```python
class DQNAgent:
    def __init__(self,num_actions,gamma=0.99,epsilon=1.0):
        self.gamma = gamma
        self.epsilon = epsilon
        self.replay_buffer = ReplayBuffer(capacity=6000)
        self.num_actions = num_actions
        self.steps = 0 #noof training steps

        self.q_net = QNetwork(num_actions).to(device)
        self.target_net = QNetwork(num_actions).to(device)
        for param in self.target_net.parameters():
          param.requires_grad = False
        self.optimizer = optim.Adam(self.q_net.parameters(), lr=0.0001)
        self.criterion = nn.MSELoss()

    def select_action(self, state):
        if np.random.rand() < self.epsilon:
            return np.random.choice(self.num_actions)
        with torch.no_grad():
            state = torch.from_numpy(state).float().unsqueeze(0).to(device)
            q_values = self.target_net(state)
            return q_values.argmax().item()

    def train_model(self, states_batch, actions_batch, rewards_batch, next_states_batch):

        state_tensor = torch.tensor(states_batch, dtype=torch.float32).to(device)
        action_tensor = torch.tensor(actions_batch, dtype=torch.long).to(device)
        reward_tensor = torch.tensor(rewards_batch, dtype=torch.float32).to(device)
        next_state_tensor = torch.tensor(next_states_batch, dtype=torch.float32).to(device)

        target = reward_tensor + self.gamma * self.target_net(next_state_tensor).max(1)[0].detach()
        predicted = self.q_net(state_tensor).gather(1, action_tensor.unsqueeze(1)).squeeze(1)
        loss = self.criterion(predicted, target)

        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

        loss_value = loss.item()
        self.steps += 1
        return loss_value
```

```python
def save_model(agent,filename):
    checkpoint = {
        'agent_epsilon': agent.epsilon,
        'steps':agent.steps,
        'q_net_state_dict': agent.q_net.state_dict(),
        'target_net_state_dict': agent.target_net.state_dict(),
        'optimizer_state_dict': agent.optimizer.state_dict(),
    }
    torch.save(checkpoint, filename)
    print(f'Model saved')
    return
```

```python
def load_model(agent,filename):
    checkpoint = torch.load(filename)
    agent.epsilon = checkpoint['agent_epsilon']
    agent.steps = checkpoint['steps']
    agent.q_net.load_state_dict(checkpoint['q_net_state_dict'])
    agent.target_net.load_state_dict(checkpoint['target_net_state_dict']),
    agent.optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    print(f'Model loaded')
    return
```

```python
import csv

def save_lists_to_csv(list_rewards, list_loss, filename):
    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows([list_rewards, list_loss])
    print('saved')
    return

def load_lists_from_csv(filename):
    list_rewards, list_loss = [], []
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        data = list(reader)
        list_rewards, list_loss = data
        print('loaded')
    return list_rewards,list_loss
```

```python
agent = DQNAgent(num_actions = action_size)
num_episodes = 301
batch_size = 32
list_rewards = []
list_loss = []
```

```python
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
load_model(agent,'/content/my_data/checkpoint_1100.pth.tar')
list_rewards,list_loss = load_lists_from_csv('/content/my_data/lists.csv')
```

```python
start = len(list_rewards)
print(start)
for episode in range(start,start + num_episodes):
    state,_ = env.reset()
    total_reward = 0
    count = 0
    count_train = 0
    terminated = False
    truncated = False
    loss_accumulated = 0.0
    store_state_four = deque(maxlen=4)
    store_prevstate_four = deque(maxlen=4)

    update(store_state_four,state)
    while not (terminated or truncated):
        if len(store_state_four) < 4:
              action = env.action_space.sample()
        else:
              action = agent.select_action(combine(store_state_four))
        new_state,reward,terminated ,truncated ,_ = env.step(action)
        total_reward += reward
        store_prevstate_four.append(store_state_four[-1])
        update(store_state_four, new_state)

        #adding into buffer
        if len(store_prevstate_four) == 4 and len(store_state_four) == 4:
          agent.replay_buffer.insert(combine(store_prevstate_four),action,reward,combine(store_state_four))

        #sending to training
        if len(agent.replay_buffer.memory) > batch_size and (count % 2 == 0):
            transitions = agent.replay_buffer.sample(batch_size)

            states_batch, actions_batch, rewards_batch, next_states_batch = zip(*transitions)
            states_batch = np.array(states_batch)
            actions_batch = np.array(actions_batch)
            rewards_batch = np.array(rewards_batch)
            next_states_batch = np.array(next_states_batch)

            loss = agent.train_model(states_batch, actions_batch, rewards_batch, next_states_batch)
            count_train += 1
            agent.epsilon = max(0.015,agent.epsilon - 0.00002)
            loss_accumulated += loss

        #copying states
        if agent.steps % 50 == 0 :
          agent.target_net.load_state_dict(agent.q_net.state_dict())
        count += 1
    list_rewards.append(total_reward)
    loss_accumulated = loss_accumulated/count_train
    list_loss.append(loss_accumulated)
    print(f"Episode: {episode}, Total Reward: {total_reward}, Epsilon: {agent.epsilon},count : {count},loss : {loss_accumulated}")
    if episode % 100 == 0:
        save_model(agent, f'/content/my_data/checkpoint_{episode}.pth.tar')
        save_lists_to_csv(list_rewards,list_loss,'/content/my_data/lists.csv')
        source_directory = '/content/my_data/'
        destination_directory = '/content/drive/MyDrive/my_data/'
        shutil.copytree(source_directory, destination_directory,dirs_exist_ok=True)

        load_model(agent, f'/content/my_data/checkpoint_{episode}.pth.tar')
        list_rewards,list_loss = load_lists_from_csv('/content/my_data/lists.csv')
```
