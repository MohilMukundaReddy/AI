### Data preprocessing

#### missing values

```python
missing_percentage = (data.isnull().sum() / len(data)) * 100
columns_with_missing = missing_percentage[missing_percentage > 0]
print("Columns with missing values and their corresponding percentage:")
print(columns_with_missing)
```

```python
data.emp_title.nunique() #since there is just 5% data just remove the rows
```

```python
#from the above emp_length all have almost same information about loan_status so not useful for classification
data.drop('emp_length', axis=1, inplace=True) 
# redundant column of purpose
data.drop('title', axis=1, inplace=True)
# there are too many categories for this categorical variable can be removes
data.drop('emp_title', axis=1, inplace=True)
```

```python
#revol_util ,pub_rec_bankruptcies have very less missing values at this point
data.dropna(subset=['revol_util', 'pub_rec_bankruptcies'], inplace=True)
```

```python
#we have around 9.5% missing vales for mort_acc but we can see it is slightly coorelated with total_acc which makes sense
#we can predict using simple assumptions like the E[mort_acc|total_acc] as follows
mean_mort_acc = data.groupby(by='total_acc').mean().mort_acc
mean_mort_acc
```

```python
data['mort_acc'] = data.apply(lambda row: mean_mort_acc[row['total_acc']] if pd.isnull(row['mort_acc']) else row['mort_acc'], axis=1)
```

#### categorical variables

```python
#grade is redundant if we have subgrade
data.drop('grade', axis=1, inplace=True)  
# issue_d is not available prior so we cant have it 
data.drop('issue_d', axis=1, inplace=True)
```

```python
#address is too unique to predict from so take pincode from it
data['address'] = data.address.apply(lambda x: x[-5:])
```

```python
#as this is a numerical in str conver to numerical
data['term'].replace({' 36 months': 36, ' 60 months': 60}, inplace=True)
```

```python
data['earliest_cr_line'] = pd.to_datetime(data['earliest_cr_line']).dt.year
```

```python
#ordinal encoding for sub grades
sub_grade_mapping = {'A1': 1, 'A2': 2, 'A3': 3, 'A4': 4, 'A5': 5,
                     'B1': 6, 'B2': 7, 'B3': 8, 'B4': 9, 'B5': 10,
                     'C1': 11, 'C2': 12, 'C3': 13, 'C4': 14, 'C5': 15,
                     'D1': 16, 'D2': 17, 'D3': 18, 'D4': 19, 'D5': 20,
                     'E1': 21, 'E2': 22, 'E3': 23, 'E4': 24, 'E5': 25,
                     'F1': 26, 'F2': 27, 'F3': 28, 'F4': 29, 'F5': 30,
                     'G1': 31, 'G2': 32, 'G3': 33, 'G4': 34, 'G5': 35}
data['sub_grade'] = data['sub_grade'].map(sub_grade_mapping)
```

```python
# Define columns for one-hot encoding
one_hot = ['verification_status','address', 'purpose', 'initial_list_status', 'application_type', 'home_ownership']
data = pd.get_dummies(data, columns=one_hot, drop_first=True)
```

```python
data['loan_status'].replace({'Fully Paid': 1, 'Charged Off': 0}, inplace=True)
```

<center><h3>Models</h3></center>

```python
def plot_roc_and_confusion_matrix(model, x_test, y_test):
    y_pred_prob = model.predict_proba(x_test)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    auc_score = roc_auc_score(y_test, y_pred_prob)
    plt.figure(figsize=(8,4))
    plt.subplot(1, 2, 1)
    plt.plot(fpr, tpr, label=f'ROC curve (AUC = {auc_score:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')

    y_pred = model.predict(x_test)
    conf_mat = confusion_matrix(y_test, y_pred)
    plt.subplot(1, 2, 2)
    sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix')   
    plt.tight_layout()
    plt.show()

    report = classification_report(y_test, y_pred,output_dict=True)
    print(classification_report(y_test, y_pred))
    return report
```

### Logistic regression with L2 regularization

```python
C_values = [0.01,0.1,1,5,10]  

mean_scores = []
for C in C_values:
    log_reg_regularized = LogisticRegression(C=C, penalty='l2', solver='liblinear')
    scores = cross_val_score(log_reg_regularized, x_train, y_train, cv=5)
    mean_scores.append(np.mean(scores))
best_index = np.argmax(mean_scores)
best_C = C_values[best_index]
```

```python
print(best_C)
```

```python
log_reg_regularized = LogisticRegression(C=best_C, penalty='l2', solver='liblinear')
log_reg_regularized.fit(x_train, y_train)
_ = plot_roc_and_confusion_matrix(log_reg_regularized, x_train, y_train)
```

```python
report1 = plot_roc_and_confusion_matrix(log_reg_regularized, x_test, y_test)
```

### Comparing models

```python
def calculate_metrics(report):
    metrics_dict = {
        'accuracy': report['accuracy'],
        'precision_class_0': report['0']['precision'],
        'precision_class_1': report['1']['precision'],
        'recall_class_0': report['0']['recall'],
        'recall_class_1': report['1']['recall']
    }
    return metrics_dict
    
def compare_class_metrics(reports, model_names):
    metrics = {
        'precision_class_0': [],
        'precision_class_1': [],
        'recall_class_0': [],
        'recall_class_1': [],
        'accuracy': []
    }
    
    for report in reports:
        metrics_dict = calculate_metrics(report)
        for metric, value in metrics_dict.items():
            metrics[metric].append(value)

    fig_acc, ax_acc = plt.subplots(figsize=(8,4))
    ax_acc.bar(model_names, metrics['accuracy'])
    ax_acc.set_title('Accuracy')
    ax_acc.set_xlabel('Model')
    ax_acc.set_ylabel('Accuracy')

    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,6))
    axes = axes.flatten()
    metrics_names = list(metrics.keys())
    
    for i in range(len(metrics_names)):
        if metrics_names[i] != 'accuracy': 
            ax = axes[i]
            metric_name = metrics_names[i]
            ax.bar(model_names, metrics[metric_name])
            ax.set_title(metric_name)
            ax.set_xlabel('Model')
            ax.set_ylabel(metric_name)
            
    plt.tight_layout()
    plt.show()
```

```python
compare_class_metrics([report1, report2, report3], ['logistic_reg', 'random_forest','XGBoost'])
```
